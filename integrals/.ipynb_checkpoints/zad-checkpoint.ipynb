```python
from numpy.polynomial.legendre import leggauss
import scipy.integrate.quadrature as qt
from scipy.integrate import quad
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
def P(i):
    if i == 0:
        return lambda x: 1 + x*0
    elif i == 1:
        return lambda x: x
    else:
        return lambda x: (2*i-1)/i*x*P(i-1)(x)-(i-1)/i*P(i-2)(x)
```

```python
t = np.arange(-1.0, 1.0, 0.01)
s = P(0)(t)
line, = plt.plot(t, s, lw=2)
```

```python
t = np.arange(-1.0, 1.0, 0.01)
s = P(1)(t)
line, = plt.plot(t, s, lw=2)
```

```python
t = np.arange(-1.0, 1.0, 0.01)
s = P(2)(t)
line, = plt.plot(t, s, lw=2)
```

```python
s = P(2)
print(scipy.optimize.bisect(s, -1, 0))
print(scipy.optimize.bisect(s, 0, 1))
```

```python
t = np.arange(-1.0, 1.0, 0.01)
s = P(3)(t)
line, = plt.plot(t, s, lw=2)
```

```python
s = P(3)
print(scipy.optimize.bisect(s, -1, -0.5))
print(scipy.optimize.bisect(s, -0.5, 0.5))
print(scipy.optimize.bisect(s, 0.5, 1))
```

```python
t = np.arange(-1.0, 1.0, 0.01)
s = P(4)(t)
line, = plt.plot(t, s, lw=2)
```

```python
s = P(4)
print(scipy.optimize.bisect(s, -1, -0.75))
print(scipy.optimize.bisect(s, -0.75, 0))
print(scipy.optimize.bisect(s, 0, 0.75))
print(scipy.optimize.bisect(s, 0.75, 1))
```

```python
def integr(k, f):
    nodes, w = leggauss(k)
    nodes = np.insert(nodes, 0, -1)
    nodes = np.append(nodes, 1)
    suma = 0
    for (i, j) in zip(nodes, nodes[1:]):
        suma += (f(i)+f(j))/2*(j - i)
    return suma
```

```python
integr(20, lambda x: 4 + 0*x)
```

```python
integr(20, lambda x: x-3)
```

```python
integr(20, lambda x: 3*x**2+x-3)
```

```python
def integ(a, b, k, f):
    g = lambda x: f((b+a)/2 + (b-a)/2*x)*(b-a)/2
    return integr(k, g)
```

```python
integ(2, 4, 20, lambda x: 4 + 0*x)
```

```python
integ(-2, 4, 20, lambda x: x-3)
```

```python
integ(0, 6, 20, lambda x: 3*x**2+x-3)
```

```python
qt(lambda x: 3*x**2+x-3, 0, 6)
```

```python
quad(lambda x: 1/np.sqrt(2*np.pi)*np.exp(-(x*x)/2), -np.inf, np.inf)
```

```python
def rec_int(f, a, b, n):
    suma = 0
    d = b-a
    xs = list(range(a*n, b*n+d, d))
    for (x, y) in zip(xs, xs[1:]):
        suma += (f(x/n)+f(y/n))/2
    return d*suma/n
```

```python
rec_int(lambda x: 5*x**4+15*x**2-4*x+1, -2, 10, 1000000)
```

```python
iters = list(x for x in range(1, 100, 1))
exact = 104892.0
a = -2
b = 10
f = lambda x: 5*x**4+15*x**2-4*x+1
ress = list(map(lambda x: abs(exact-rec_int(f, a, b, x)), iters))
plt.plot(list(map(lambda x: str(x), iters)), ress, 'r')
plt.show()
```
