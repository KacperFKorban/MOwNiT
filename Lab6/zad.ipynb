```python
import pandas as pd

df = pd.read_csv('kc_house_data.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
ax = df['grade'].plot.hist()
```

```python
df['yr_renovated'] = list(zip(df['yr_renovated'], df['yr_built']))
df['yr_renovated'] = df['yr_renovated'].apply(lambda x: max(x[0], x[1]))
df.head()
```

```python
df = df.drop(['lat', 'long'], axis=1)
df.head()
```

```python
df['since_renovation'] = list(zip(df['yr_renovated'], df['date']))
df['since_renovation'] = df['since_renovation'].apply(lambda x: int(x[1][:4]) - x[0])
df.head()
```

```python
df['age'] = list(zip(df['yr_built'], df['date']))
df['age'] = df['age'].apply(lambda x: int(x[1][:4]) - x[0])
df.head()
```

```python
df['date'] = df['date'].apply(lambda x: x[:8])
df.head()
```

```python
df = df.drop(['yr_built', 'yr_renovated', 'zipcode'], axis=1)
df.head()
```

```python
df = df.drop('id', axis=1)
df.head()
```

```python
ax = df['price'].plot.hist()
```

```python
for i in range(100):
    df['price'] = df['price'].loc[df['price'] != df['price'].max()]
df['price'].plot.hist()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
prices = df['price']
prices
```

```python
df = df.drop(columns='price')
```

```python
from sklearn import preprocessing
df = preprocessing.scale(df)
df = pd.DataFrame(df, columns=['date','bedrooms','bathrooms','sqft_living','sqft_lot','floors','waterfront','view','condition','grades','qft_above','sqft_basement','sqft_living15','sqft_lot15','since_renovation','age'])
df.head()
```

```python
s = df.shape[0]
s
train1 = df.take(list(range(0, s//3)))
train2 = df.take(list(range(s//3, 2*s//3)))
train3 = df.take(list(range(2*s//3, s)))
prices1 = prices.take(list(range(0, s//3)))
prices2 = prices.take(list(range(s//3, 2*s//3)))
prices3 = prices.take(list(range(2*s//3, s)))
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(train1, prices1)
reg.coef_
```

```python
rid = linear_model.Ridge()
rid.fit(train2, prices2)
rid.coef_
```

```python
las = linear_model.Lasso()
las.fit(train3, prices3)
las.coef_
```

```python
import matplotlib.pyplot as plt
regl = [reg.score(train2, prices2), reg.score(train3, prices3)]
ridl = [rid.score(train1, prices1), rid.score(train3, prices3)]
lasl = [las.score(train1, prices1), las.score(train2, prices2)]
plt.plot([2, 3], regl, "ro", label='Regular')
plt.plot([1, 3], ridl, "bo", label='Ridge')
plt.plot([1, 2], lasl, "go", label='Lasso')
plt.legend(loc='upper right')
plt.title('Score')
plt.show()
```
