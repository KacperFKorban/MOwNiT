```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

sigma, beta, rho = 10, 2.667, 28
x0, y0, z0 = 0, 1, 1.05

tmax, n = 100, 10000

def lorenz(V, t, sigma, beta, rho):
    x, y, z = V
    xp = -sigma*(x - y)
    yp = rho*x - y - x*z
    zp = -beta*z + x*y
    return xp, yp, zp

t = np.linspace(0, tmax, n)
f = odeint(lorenz, (x0, y0, z0), t, args=(sigma, beta, rho))
x, y, z = f.T

fig = plt.figure()
ax = fig.gca(projection='3d')

s = 1000
c = np.linspace(0,1,n)
for i in range(0,n-s,s):
    ax.plot(x[i:i+s+1], y[i:i+s+1], z[i:i+s+1], color=(0, c[i], 1), alpha=0.4)

plt.show()
```
