```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from scipy.linalg import svd
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = np.outer(np.cos(u), np.sin(v))
y = np.outer(np.sin(u), np.sin(v))
z = np.outer(np.ones(np.size(u)), np.cos(v))

sph = np.array([x, y, z])

ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_zlim(-2, 2)
ax.plot_surface(sph[0], sph[1], sph[2], color='b')

plt.show()
```

```python
def transformed_sphere(A):
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    x = np.outer(np.cos(u), np.sin(v)) * A.item((0, 0)) + np.outer(np.sin(u), np.sin(v)) * A.item((0, 1)) + np.outer(np.ones(np.size(u)), np.cos(v)) * A.item((0, 2))
    y = np.outer(np.cos(u), np.sin(v)) * A.item((1, 0)) + np.outer(np.sin(u), np.sin(v)) * A.item((1, 1)) + np.outer(np.ones(np.size(u)), np.cos(v)) * A.item((1, 2))
    z = np.outer(np.cos(u), np.sin(v)) * A.item((2, 0)) + np.outer(np.sin(u), np.sin(v)) * A.item((2, 1)) + np.outer(np.ones(np.size(u)), np.cos(v)) * A.item((2, 2))
    return np.array([x, y, z])
    
A = np.matrix([[3,1,0],[1,2,0],[1,0,1]])
sph = transformed_sphere(A)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_zlim(-2, 2)
ax.plot_surface(sph[0], sph[1], sph[2], color='b')

plt.show()
```

```python
U, s, VT = svd(A)
print(U)
print(s)
print(VT)
```

```python
Sigma = np.diag(s)
B = U.dot(Sigma.dot(VT))
print(B)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for l in U:
    ax.quiver(0, 0, 0, l[0], l[1], l[2], color='r')
# ax.plot_surface(sph[0], sph[1], sph[2], color='b')
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_zlim(-2, 2)
plt.show()
```
